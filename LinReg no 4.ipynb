```python
years = df[['YearsExperience']]
salary = df[['Salary']]
years_train, years_test, salary_train , salary_test = ms.train_test_split(years,salary, test_size=0.2)
```

```python
years_train.shape
```

```python
model1 = ln.LinearRegression()
model1.fit(years_train,salary_train)
```

```python
score1 = model1.score(years_test,salary_test)
score1
```
